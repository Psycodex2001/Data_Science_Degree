Exercício 01 - Análise Univariada

```python
import os.path
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from scipy import stats

%matplotlib inline

# Tabela por Sexo
dt = pd.read_csv("https://raw.githubusercontent.com/cibelerusso/VED/main/Dados/dados_banco.csv")
dt.head()

```

```python
dt_counts = dt['Sexo'].value_counts().reset_index()
dt_counts.columns = ['Sexo', 'Count']
dt_counts
```

```python
# Gráfico por Sexo
plt.figure(figsize=(8, 6))
plt.pie(dt_counts['Count'], labels = dt_counts["Sexo"], autopct="%1.1f%%", startangle=90, colors= ["#0000ff", "#ff0080"])
plt.axis('equal')

plt.title('Gráfico de pizza da Contagem de Sexos')

plt.show()
```

```python
sns.set_palette('colorblind')
sns.relplot(x='Idade', y='Salario', data=dt)
```

```python
idade_mean = dt['Idade'].mean()
idade_std = dt['Idade'].std()

salario_mean = dt['Salario'].mean()
salario_std = dt['Salario'].std()

plt.figure(figsize=(8, 6))
sns.scatterplot(x='Idade', y='Salario', data=dt)

plt.axhline(y=salario_mean, color='red', linestyle='--', label='Salary Mean')
plt.axhline(y=salario_mean + salario_std, color='orange', linestyle='--', label='Salary Mean + Std Dev')
plt.axhline(y=salario_mean - salario_std, color='orange', linestyle='--', label='Salary Mean - Std Dev')
plt.axvline(x=idade_mean, color='blue', linestyle='--', label='Age Mean')

plt.title('Scatter Plot of Age vs. Salary with Trend Lines')
plt.xlabel('Age')
plt.ylabel('Salary')

plt.legend()

plt.show()
```
