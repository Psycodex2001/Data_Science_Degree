```python
import numpy as np

from google.colab import drive
drive.mount('/content/drive')

import pandas
# Caminho para o arquivo CSV após montar o Google Drive
file_path = '/content/drive/MyDrive/2_2023/Cópia de iris.csv'
# carregando iris.csv
dados = pandas.read_csv(file_path)
# imprimindo os dez primeiros exemplos da base de dados
print(dados.head(10))
```

```python
from matplotlib import pyplot

classe_cor = {'Setosa': 'red', 'Virginica': 'blue', 'Versicolor': 'green'}

cores = [classe_cor[nome] for nome in dados['variety']]

# print(dados.columns)

dados.plot(kind='scatter', x='sepal.length', y='sepal.width', c=cores)
# plot.show()
```
