1. O que é Mineração de Dados?

Mineração de dados é uma área da ciência da computação que se dedica a extrair conhecimento e informações úteis a partir de grandes conjuntos de dados. O objetivo principal da mineração de dados é descobrir padrões, tendências e relações ocultas nos dados, a fim de tomar decisões mais informadas e prever resultados futuros.

A mineração de dados utiliza técnicas e algoritmos de aprendizado de máquina, estatística e inteligência artificial para analisar os dados e identificar informações relevantes. Ela pode ser aplicada em diversas áreas, como marketing, finanças, medicina, logística e segurança, entre outras. Por exemplo, em marketing, a mineração de dados pode ser usada para identificar padrões de consumo dos clientes e segmentar o público-alvo de uma empresa com maior precisão.

Existem diferentes etapas no processo de mineração de dados:


1.   A primeira etapa é a coleta dos dados, que podem ser obtidos de diversas fontes, como bancos de dados, redes sociais, sensores e dispositivos inteligentes;
2.   A segunda, os dados são pré-processados e limpos, removendo-se ruídos e *outliers*;
3.   Em seguida, é feita uma análise exploratória dos dados, com o objetivo de entender melhor suas características e identificar possíveis relações entre as variáveis; e
4.   Por fim, são aplicadas técnicas de aprendizado de máquina e estatística para extrair conhecimento dos dados e construir modelos preditivos.

2.1.1. *Arrays*

O principal tipo de dados disponibilizado pela biblioteca *NumPy* e que será mais utilizado no decorrer da disciplina é conhecido como *numpy.array*. Um *array* pode ser instanciado por meio da chamada *numpy.array(lista)*, na qual lista é um objeto do tipo *list* contendo apenas valores numéricos.

```python
import numpy
lista = [1, 2, 3, 4, 5]
x = numpy.array(lista)
print(x)
```

O tipo *array* fornece facilidades para a realização de operações matemáticas com escalares. Para fins de ilustração, execute o código abaixo:


```python
import numpy
x = numpy.array([1, 2, 3, 4, 5])
y = 2
print(x + y)

print(x - y)
print(y - x)

print(x * y)

print(x / y)
print(y / x)
```

Ao inspecionar os resultados do código anterior, é possível observar que nas operações de soma, subtração, multiplicação e divisão em que um dos operandos é um escalar e o outro é um *array*, cada operação ocorre entre o escalar e cada elemento do *array*.

A *NumPy* também possui funções para realizar operações utilizando dois *arrays* de mesmo tamanho. Como exemplo, execute o código abaixo:


```python
import numpy
x = numpy.array([1, 2, 3, 4, 5])
y = numpy.array([6, 7, 8, 9, 10])

print(x + y)

print(x - y)
print(y - x)

print(x * y)

print(x / y)
print(y / x)
```

Ao analisar as saídas do código anterior, pode-se perceber que, quando os operandos são dois *arrays* do mesmo tamanho, as operações de soma, subtração, multiplicação e divisão ocorrem elemento a elemento. Embora os *arrays* utilizados nos exemplos anteriores tenham apenas uma dimensão, as operações podem ser aplicadas a *arrays* e matrizes com qualquer número de dimensões.

Por fim, a *NumPy* fornece ainda várias funções pré-definidas para explorar diversas propriedades de um *array* qualquer *x*. Dentre as funções que serão utilizadas ao longo da disciplina, pode-se mencionar:


*   numpy.sum(x): retorna a soma de todos os elementos de x.
*   numpy.max(x): retorna o valor máximo contido em x.
*   numpy.min(x): retorna o valor mínimo contido em x.

2.2.1.1. Medidas de posição ou localidade

2.2.1.1.1 Média

Considerando a existência de *n* observações diferentes para uma variável qualquer *x*, a média é definida como:

média.png

Na biblioteca *NumPy*, assumindo que um conjunto de *n* observações seja representado por um *array* de tamanho *n*, a média pode ser calculada conforme o exemplo a seguir:


```python
import numpy
# gerando um array com 100 valores aleatorios
# sorteados entre 0 e 1000
x = numpy.random.randint(low=0, high=100, size=100)
# calculando a media utilizando os metodos
# numpy.sum e len
media = numpy.sum(x) / len(x)
print(media)
# mais facilmente, podemos utilizar o metodo numpy.mean
media = numpy.mean(x)
print(media)
```

2.2.1.1.2 Mediana

Considerando a existência de n observações para uma variável qualquer *x*, a mediana pode ser definida como o valor central das observações ordenadas quando *n* for ímpar. Para os casos em que *n* for par, a mediana é definida como valor médio entre as duas observações centrais ordenadas [Faceli et al. 2011]. Em Python, a mediana pode ser calculada tal
como apresentado no exemplo abaixo:

```python
mediana = numpy.median(x)
print(mediana)
```

2.2.1.1.3 Percentis

Um percentil ou quantil, denotado por *py%*, representa, dentre *m* observações, o valor tal que *y%* das observações são menores do que ele [Magalhães e de Lima 2000, Faceli *et al*. 2011]. Esta medida trata-se de uma generalização da mediana (que corresponde a *p50%*) e do primeiro e terceiro quartis (que correspondem a *p25%* e *p75%*, respectivamente). Para o seu cálculo, a biblioteca *NumPy* dispõe do seguinte método:


```python
# para calcular p15%
percentil_15 = numpy.percentile(x, 15)
print(percentil_15)
```

2.2.1.2. Medidas de dispersão ou espalhamento

2.2.1.2.1 Intervalo ou amplitude

O intervalo (também conhecido como amplitude) consiste na diferença entre o valor máximo e mínimo de um conjunto de observações. Em Python, o mesmo pode ser calculado
como:

```python
valor_maximo = numpy.max(x)
valor_minimo = numpy.min(x)
intervalo = valor_maximo - valor_minimo
print(intervalo)
```

2.2.1.2.2 Variância e desvio-padrão

A variância de um conjunto de observações, denotada por *s2*, é definida como:

variança.png

em que ¯x consiste na média das observações (Equação 1). Para manter a mesma unidade dos dados originais, o desvio-padrão é definido como [Magalhães e de Lima 2000]:

media das obs.png


O cálculo dessas duas medidas pode ser implementado como:

```python
media = numpy.mean(x)
n = len(x)
diferencas = x - media
variancia = numpy.sum(diferencas * diferencas) / (n - 1)
desvio_padrao = numpy.sqrt(variancia)
# de maneira mais facil, podemos utilizar as seguintes funcoes
variancia = numpy.var(x, ddof=1)
desvio_padrao = numpy.std(x, ddof=1)
```

em que o parâmetro *ddof* indica o número de graus de liberdade utilizados. Para o cálculo na Equação, as funções *numpy.var* e *numpy.std* utilizam como divisor a fórmula *n − ddof*. Em todos os cenários estudados neste material será utilizada a variância amostral e o desvio-padrão amostral. Portanto, o valor considerado para ddof sempre será 1.

2.2.1.3. Medidas de distribuição

2.2.1.3.1 Obliquidade

A obliquidade (ou skewness, em inglês) trata-se de uma medida de assimetria de uma distribuição de probabilidade em torno de sua média. Ela pode assumir valores negativos, positivos ou próximos de 0.

O cálculo da obliquidade é definido por:

obliquidade.png

sendo *¯x* e *s* definidos tal como nas equações.

Em Python, a obliquidade dos dados contidos em um array pode ser calculada por meio da biblioteca *SciPy* da seguinte maneira:

```python
import numpy
import scipy.stats
# gerando uma amostra com 10000 observacoes a partir de
# uma distribuicao normal com media zero e desvio-padrao
# unitario
dados = numpy.random.normal(loc=0.0, scale=1.0, size=10000)
# como os dados foram gerados segundo uma distribuicao
# normal, que eh simetrica, a obliquidade devera resultar
# em algum valor proximo de 0
obliquidade = scipy.stats.skew(dados)
print(obliquidade)
```

2.2.1.3.2 Curtose

A curtose é uma medida que caracteriza o achatamento da distribuição dos dados. Assim como a obliquidade, os seus valores podem ser negativos, positivos ou próximos de 0.

A equação para o cálculo da curtose é definida como:

curtose.png

sendo *¯x* e *s* definidos, respectivamente.

Tal como a obliquidade, a curtose pode ser calculada por meio da biblioteca SciPy:

```python
import numpy
import scipy.stats
# gerando uma amostra com 10000 observacoes a partir de
# uma distribuicao normal com media zero e desvio-padrao
# unitario
dados = numpy.random.normal(loc=0.0, scale=1.0, size=10000)
# como os dados foram gerados segundo uma distribuicao
# normal, que eh simetrica, a curtose devera resultar
# em algum valor proximo de 0
curtose = scipy.stats.kurtosis(dados)
print(curtose)
```

2.4. Visualização de dados

2.4.1. Histogramas

Uma das formas mais simples de ilustrar a distribuição de um conjunto de valores de uma variável é o uso de histogramas. Neste tipo de gráfico tem-se, no eixo horizontal, o conjunto (ou intervalos) de valores observados, enquanto que no eixo vertical, apresentase a frequência de ocorrência de cada valor (ou valores dentre de um intervalo) presente na amostra analisada. O pacote NumPy fornece uma função para calcular o histograma, que pode ser vista abaixo. Nessa função, bins corresponde ao número de barras verticais. Quando o valor de bins é definido como *auto*, o número de barras é definido
automaticamente. Como exemplo, considere o código a seguir:

```python
# calculando o histograma para uma variável
import numpy
from matplotlib import pyplot
# Notas na primeira prova
notas = numpy.array([2, 5, 7, 3, 5, 6, 5, 6, 6, 5, 5, 3])
pyplot.hist(notas, bins='auto')
pyplot.title('Histograma')
pyplot.ylabel('Frequencia')
pyplot.xlabel('Nota')
pyplot.show()
```

2.4.2. Scatter plots

Um scatter plot consiste em um tipo de gráfico comumente utilizado para observar o comportamento entre duas variáveis de uma base de dados.

```python
import pandas as pd
from matplotlib import pyplot as plt

# Defina o caminho correto para o arquivo iris.csv no seu Google Drive
file_path = '/content/drive/MyDrive/2_2023 - Mineração de Dados/Aula 1 - Introdução à Mineração de Dados/iris.csv'

# Carregando iris.csv
dados = pd.read_csv(file_path)

# Criando um dicionário para mapear cada classe para uma cor
classe_cor = {'Setosa': 'red', 'Virginica': 'blue', 'Versicolor': 'green'}

# Criando uma lista com as cores de cada exemplo
# Supondo que a coluna com as classes seja 'variety'
cores = [classe_cor[nome] for nome in dados['variety']]

# Verificando os nomes das colunas
print(dados.columns)

# Gerando scatter plot
# No eixo x será plotado o tamanho da sepala
# No eixo y será plotado o comprimento da sepala
dados.plot(kind='scatter', x='sepal.length', y='sepal.width', c=cores)
plt.show()
```

Alternativamente, pode-se também gerar uma matriz de scatter plots, que irá conter os scatter plots para todos os pares possíveis de atributos. Ademais, na diagonal de tal matriz, pode-se apresentar informações a respeito de cada atributo (por exemplo, o seu histograma).

```python
import pandas
from pandas.plotting import scatter_matrix
from matplotlib import pyplot

# Defina o caminho correto para o arquivo iris.csv no seu Google Drive
file_path = '/content/drive/MyDrive/2_2023 - Mineração de Dados/Aula 1 - Introdução à Mineração de Dados/iris.csv'

# Carregando iris.csv
dados = pd.read_csv(file_path)


# criando um dicionario para mapear cada classe para uma cor
classe_cor = {'Setosa' : 'red', 'Virginica' : 'blue', 'Versicolor' : 'green'}

# criando uma lista com as cores de cada exemplo
cores = [classe_cor[nome] for nome in dados.variety]

# gerando matriz de scatter plots
scatter_matrix(dados, color=cores)
pyplot.show()
```

2.4.3. *Box plots*

Um box plot é um gráfico apresentado em formato de caixa, em que a aresta inferior da caixa representa o primeiro quartil (Q1), a aresta superior representa o terceiro quartil (Q3) e um traço interno à caixa representa a mediana (Q2) de uma amostra. Ademais, uma linha tracejada delimita o limite entre o terceiro quartil e o maior valor das observações que é menor ou igual a Q3 +1,5 ·(Q3 − Q1) e o limite entre o primeiro quartil e o menor valor na amostra que é maior ou igual a Q1 − 1,5 ·(Q3 − Q1).

```python
import pandas
from pandas.plotting import scatter_matrix
from matplotlib import pyplot

# Defina o caminho correto para o arquivo iris.csv no seu Google Drive
file_path = '/content/drive/MyDrive/2_2023 - Mineração de Dados/Aula 1 - Introdução à Mineração de Dados/iris.csv'

# Carregando iris.csv
dados = pd.read_csv(file_path)

# gerando bloxplot para os atributos da Iris
dados.plot(kind='box')
pyplot.show()
```
