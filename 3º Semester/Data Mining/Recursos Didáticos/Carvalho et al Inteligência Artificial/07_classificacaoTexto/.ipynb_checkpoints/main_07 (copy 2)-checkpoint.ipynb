## Classificação de spam

Nesta parte do exercício, usaremos o *naive* Bayes multinomial para classificar SMS spam.

Abaixo são mostrados alguns exemplos de SMS legítimos:
 * ```Is that seriously how you spell his name?```
 * ```What you thinked about me. First time you saw me in class.```
 * ```Ok lar i double check wif da hair dresser already he said wun cut v short. He said will cut until i look nice.```
 
Abaixo são mostrados alguns exemplos de SMS spam:
 * ```WINNER!! As a valued network customer you have been selected to receivea £900 prize reward! To claim call 09061701461. Claim code KL341. Valid 12 hours only.```
 * ```Thanks for your subscription to Ringtone UK your mobile will be charged £5/month Please confirm by replying YES or NO. If you reply NO you will not be charged.```
 * ```Congratulations ur awarded 500 of CD vouchers or 125gift guaranteed & Free entry 2 100 wkly draw txt MUSIC to 87066 TnCs http://www.Ldew.com1win150ppmx3age16.```

Dentro da pasta *datasets*, há uma base de dados de SMS. Essa base de dados também pode ser encontrada no seguinte link: <https://archive.ics.uci.edu/ml/datasets/SMS+Spam+Collection>.

Vamos importar a base de dados. 

````python
def import_dataset(pathDataset):
    """
    Função usada para importar a base de dados textual
    
    Parameters:
    -----------
    pathDataset: string
        É o endereço da base de dados. Cada linha da base de dados, deve ter o formato <classe, mensagem>.        
    """
     
    datasetFile = open(pathDataset,'r') #abre arquivo para leitura
    
    dataset = [] #lista vazia que ira guardar as mensagens
    target = []
    for line in datasetFile:
        splitLine = line.split(',')
        
        classe = splitLine[0] 
        texto = ' '.join(splitLine[1::])
        
        target.append(classe)
        dataset.append(texto)
        
        #print('%s - %s' %(classe,texto))
    
    target = np.asarray(target) #convert a lista para uma array do numpy
    dataset = np.asarray(dataset) #convert a lista para uma array do numpy
        
    return dataset, target

# importa a base de dados
dataset, target = import_dataset('datasets/SMSSpamCollection.txt')

print('5 primeiras amostras: \n')
print(dataset[0:5])
````

Vamos dividir os dados em treino e teste de forma estratificada. Neste exercício iremos usar a validação cruzada holdout estratificada por motivos didáticos. No entanto, sempre que for possível, é mais recomendado usar a validação cruzada K-fold, pois é mais robusta.

````python
# gera 1 divisão dos dados em treino e teste
cv = skl.model_selection.StratifiedShuffleSplit(n_splits=1, test_size=0.3, random_state=2019)

# retorna os índices de treino e teste
train_index, test_index = list( cv.split(dataset, target) )[0]

# retorna as partições de treino e teste de acordo com os índices
dataset_train, dataset_test = dataset[train_index], dataset[test_index];
Y_train, Y_test = target[train_index], target[test_index];

print('Qtd. dados de treinamento: %d (%1.2f%%)' %(dataset_train.shape[0], (dataset_train.shape[0]/dataset.shape[0])*100) )
print('Qtd. de dados de teste: %d (%1.2f%%)' %(dataset_test.shape[0], (dataset_test.shape[0]/dataset.shape[0])*100) )

# imprime a porcentagem de dados de treinamento de cada classe
print("\nQtd. de dados de cada classe (treinamento)")
cTrain, counts_cTrain = np.unique(np.sort(Y_train), return_counts=True)
for i in range( len(cTrain) ):
    print('\tClasse %s: %d (%1.2f%%)' %( cTrain[i],counts_cTrain[i],(counts_cTrain[i]/len(Y_train))*100 ) )

# imprime a porcetagem de dados de teste de cada classe
print("\nQtd. de dados de cada classe (teste)")
cTest, counts_cTest = np.unique(np.sort(Y_test), return_counts=True)
for i in range( len(cTrain) ):
    print('\tClasse %s: %d (%1.2f%%)' %( cTest[i],counts_cTest[i],(counts_cTest[i]/len(Y_test))*100 ) )
````

Antes de fazer qualquer tarefa de classificação com textos, é importante fazer um pré-processamento para obter melhor resultado na predição. 

**Importante:** as técnicas de tratamentos de texto que vamos aplicar nesse exercício não são as únicas. Além disso, uma técnica pode ajudar a classificação em um determinado problema e não ajudar ou até mesmo atrapalhar a classificação em outro problema. 


Na função abaixo, os seguintes pré-processamentos são realizados:

 - deixar todas as palavras com letras minúsculas
 - substituir os números pela palavra *number*
 - substituir todas as URLS pela palavra *enderecoweb*
 - substiuir todos os emails pela palavra *enderecoemail*
 - substituir o símbolo de dólar pela palavra *dolar*
 - substituir todos os caracteres não-alfanuméricos por um espaço em branco
 
Além dos pré-processamentos apresentados acima, nós também podemos aplicar *estemização*. A estemização diminui as palavras por seus radicais. Veja como fica algumas palavras após a estemização: 
 - *flies* é substituída por *fli*
 - *sensational* é substituída por *sensat*
 - *plotted* é subsituída por plot
 
Para aplicar a estemização iremos usar uma função da biblioteca **NLTK** (https://www.nltk.org/). Para que a função de estemização funcione, você precisa fazer instalar a biblioteca e fazer o download dos seus pacotes, usando o script ***nltk.download('all')***. 
 
Por fim, também é recomendado eliminar todas as palavras muito curtas. Vamos eliminar qualquer palavra de apenas 1 caracter. 

````python
import nltk
from nltk.stem.porter import PorterStemmer #para fazer a estemização

def preprocessing(text):
    
    # Lower case
    text = text.lower()
    
    # remove tags HTML
    regex = re.compile('<[^<>]+>')
    text = re.sub(regex, " ", text) 

    # normaliza os numeros 
    regex = re.compile('[0-9]+')
    text = re.sub(regex, "number", text)
    
    # normaliza as URLs
    regex = re.compile('(http|https)://[^\s]*')
    text = re.sub(regex, "enderecoweb", text)

    # normaliza emails
    regex = re.compile('[^\s]+@[^\s]+')
    text = re.sub(regex, "enderecoemail", text)
    
    #normaliza o símbolo de dólar
    regex = re.compile('[$]+')
    text = re.sub(regex, "dolar", text)
    
    # converte todos os caracteres não-alfanuméricos em espaço
    regex = re.compile('[^A-Za-z]') 
    text = re.sub(regex, " ", text)
    
    # substitui varios espaçamentos seguidos em um só
    text = ' '.join(text.split())
    
    # aplica estemização
    stemmer_method = PorterStemmer()  
    words = text.split() # separa o texto em palavras
    words = [ stemmer_method.stem(w) for w in words ]
    text = " ".join( words )
        
    return text

smsContent = 'Congratulations ur awarded 500 of CD vouchers or 125gift guaranteed & Free entry 2 100 wkly draw txt MUSIC to 87066 TnCs http://www.Ldew.com1win150ppmx3age16.'
print('Antes do preprocessamento: \n', smsContent)

# chama a função de pré-processsamento para tratar o email
smsContent = preprocessing(smsContent)

print('\nDepois do preprocessamento: \n', smsContent)
````

Aplique o pré-processamento nos dados de treinamento e de teste. 

````python
# aplica o pre-processamento dos dados de treinamento
for i, msg in enumerate(dataset_train):
    dataset_train[i] = preprocessing(dataset_train[i])
    
print('Primeira mensagem de treinamento apos o pre-processamento: \n\t', dataset_train[0])
    
# aplica o pre-processamento dos dados de teste
for i, msg in enumerate(dataset_test):
    dataset_test[i] = preprocessing(dataset_test[i])
    
print('\nPrimeira mensagem de teste apos o pre-processamento: \n\t', dataset_test[0])
````

Depois de fazer o pré-processamento, é necessário transformar o texto em um vetor de atributos com valores numéricos. Uma das formas de fazer isso é considerar que cada palavra da base de dados de treinamento é um atributo que armazena o número de vezes que uma determinada palavra aparece no texto. No *scikit-learn* podemos fazer essa conversão de texto para um vetor de atributos usando a função **skl.feature_extraction.text.CountVectorizer**. Essa função gera um modelo de vetorização que pode ser ajustado com base nos dados de treinamento usando a função *fit_transform**

````python
# inicia o modelo usado para gerar a representação TF (term frequency)
vectorizer = skl.feature_extraction.text.CountVectorizer(analyzer = "word", tokenizer = None, preprocessor = None, stop_words = None, lowercase = True, binary=False, dtype=np.int32)

# treina o modelo TF com os dados de treinamento e converte os dados de treinamento para uma array que contém a frequência dos termos em cada documento (TF - term frequency)
X_train = vectorizer.fit_transform(dataset_train) 

print('20 primeiras palavras do vocabulário obtido a partir dos dados de treinamento:\n')
print(vectorizer.get_feature_names()[0:20])

print('\nDimensão dos dados de treinamento vetorizados: ', X_train.shape)

# converte os dados de teste para uma array que contém a frequência dos termos em cada documento (TF - term frequency)
X_test = vectorizer.transform(dataset_test) #converte os dados de teste, usando o modelo gerado pelos dados de treinamento 
print('\nDimensão dos dados de teste vetorizados: ', X_test.shape)
````

Como os dados de treinamento e teste já estão vetorizados, podemos usá-los para treinar e testar qualquer método de classificação. Primeiramente, vamos testar o *naive* Bayes multinomial.

````python
# carrega o classificador
NBclassifier = skl.naive_bayes.MultinomialNB() 

# treina o classificador com os dados de treinameto
NBclassifier.fit(X_train, Y_train) 
 
# classifica os dados de teste
Y_pred = NBclassifier.predict(X_test) 
        
print('\nPredição obtida para as 20 primeiras amostras de teste:\n', Y_pred[0:20])
````

Agora devemos fazer criar uma função para tratar os textos da base de dados que foi importada. As notícias da base de dados foram escritas em língua portuguesa. Portanto, devemos usar uma função de estemização apropriada para a língua portuguesa, diferindo do que foi feito nos experimentos com SMS spam. Antes disso, iremos também remover os acentos das palavras.

````python
import nltk
from nltk.stem import RSLPStemmer #portuguese

import unicodedata #remover acentos

def preprocessing_portuguese(text):
    
    # Lower case
    text = text.lower()
    
    # remove os acentos das palavras
    nfkd_form = unicodedata.normalize('NFKD', text)
    text = u"".join([c for c in nfkd_form if not unicodedata.combining(c)])
    
    # remove tags HTML
    regex = re.compile('<[^<>]+>')
    text = re.sub(regex, " ", text) 

    # normaliza os numeros 
    regex = re.compile('[0-9]+')
    text = re.sub(regex, "number", text)
    
    # normaliza as URLs
    regex = re.compile('(http|https)://[^\s]*')
    text = re.sub(regex, "enderecoweb", text)

    # normaliza emails
    regex = re.compile('[^\s]+@[^\s]+')
    text = re.sub(regex, "enderecoemail", text)
    
    #normaliza o símbolo de dólar
    regex = re.compile('[$]+')
    text = re.sub(regex, "dolar", text)
    
    # converte todos os caracteres não-alfanuméricos em espaço
    regex = re.compile('[^A-Za-z]') 
    text = re.sub(regex, " ", text)
    
    # substitui varios espaçamentos seguidos em um só
    text = ' '.join(text.split())
    
    # aplica estemização
    stemmer_method = RSLPStemmer()  
    words = text.split() # separa o texto em palavras
    words = [ stemmer_method.stem(w) for w in words ]
    text = " ".join( words )
        
    return text

exemplo_noticia = 'Polos magnéticos da Terra podem se inverter e causar colapso mundial: "A Terra ficaria inabitável". Aos menos 100 mil pessoas morreriam por ano pela alta nos níveis de radiação espacial." Se o campo magnético continuar a diminuir e os polos magnéticos se inverterem, a Terra pode acabar como Marte – um local seco, árido e incapaz de preservar a vida.'
print('Antes do preprocessamento: \n', exemplo_noticia)

# chama a função de pré-processsamento para tratar o email
exemplo_noticia = preprocessing_portuguese(exemplo_noticia)

print('\nDepois do preprocessamento: \n', exemplo_noticia)
````
