```python
from pyomo.environ import *
```

```python
# Cria um modelo concreto
model = ConcreteModel()

# Define o conjunto de produtos: "A" e "B"
model.produtos = Set(initialize=["A", "B"])
# Cria uma variável Q indexada pelo conjunto de produtos e garante valores inteiros não-negativos
model.Q = Var(model.produtos, within=NonNegativeIntegers)
```

```python
# Dicionário com os dados dos produtos
opt_prod = {
    "A": {"Tempo": 2, "Quantidade": 3, "Lucro": 30}, # Opção A: 2 horas, 3 unidades e 30 reais
    "B": {"Tempo": 3, "Quantidade": 2, "Lucro": 50}, # Opção B: 3 horas, 2 unidades e 50 reais
}
```

```python
# Variáveis de restrição
lim_tempo = 100
lim_quant = 120

# Restrição de tempo de máquina: máximo 100 horas
model.rest_time = Constraint(
    expr=sum(opt_prod[p]["Tempo"] * model.Q[p] for p in model.produtos) <= lim_tempo
)
# Restrição de matéria-prima: máximo 120 unidades
model.rest_quant = Constraint(
    expr=sum(opt_prod[p]["Quantidade"] * model.Q[p] for p in model.produtos) <= lim_quant
)

# Função objetivo com o sentido de maximizar o lucro total
model.lucro_total = Objective(expr=sum(opt_prod[p]["Lucro"] * model.Q[p] for p in model.produtos),
```

```python
# Resolve o modelo usando GLPK
solver = SolverFactory('glpk')
resultado = solver.solve(model)

# Verificar se encontrou uma solução ótima
if resultado.solver.status == SolverStatus.ok:
    print("\nSolução encontrada:\n",
          f"Quantidade do Produto A: {value(model.Q["A"])}",
          f"Quantidade do Produto B: {value(model.Q["B"])}",
          f"\nO lucro total obtido foi: R$ {value(lucro_total):.2f}")
else:
    print("Não foi possível encontrar uma solução ótima.") # Não obteve resultados...
```
