```python
import numpy as np
import pandas as pd
import csv

cotacao = []

with open('dolar.txt', encoding='utf-8') as arquivo:
  tabela = csv.reader(arquivo)
  for i in tabela:
    cotacao.append(float(i[1]))

opcao = int(input("Digite a opção desejada:\n1) Valor médio do dólar no mês de agosto.\n2) Média móvel semanal.\n3) Desvio padrão amostral da cotação mensal.\n"))

if opcao == 1:
  media_mensal = 0
  for i in range(len(cotacao)):
    media_mensal = media_mensal + cotacao[i]
  print("A média mensal de agosto foi: %.4f" %(media_mensal/len(cotacao)))

elif opcao == 2:
  d = pd.Series(cotacao)
  print("A média móvel mensal foi: \n", d.rolling(5).mean())

elif opcao == 3:
  desvio_padrao = 0
  desvio_padrao = np.std(cotacao, ddof = 1)
  print("O Desvio Padrão Amostral da cotação mensal foi: %.4f" %(desvio_padrao))


else:
  print("Opção não existente!")
```
