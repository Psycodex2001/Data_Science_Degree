```python
x = int(input("Digite a medida em pés: ")) #recebe como entrada o valor da medida em pés
conv_polegadas = x*12 #converte a medida recebida de pés para polegadas e armazena na variável conv_polegadas
conv_jardas = x/3 #converte a medida recebida de pés para jardas e armazena na variável conv_jardas
conv_milhas = conv_jardas/1760 #converte o valor armazenado em conv_jardas para milhas e armazena na variável conv_milhas
print("A medida digitada equivale a:\n%.2f polegadas.\n%.2f jardas.\n%.6f milhas." %(conv_polegadas, conv_jardas, conv_milhas)) #Apresenta no terminal de saída o valor inicial convertido em polegadas, jardas e milhas





```
