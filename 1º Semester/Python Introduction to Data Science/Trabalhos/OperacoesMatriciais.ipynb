```python
import numpy as np

# Função que é definida para criar as matrizes, recebendo como parâmetros o número de linhas e de colunas.
def criar_matriz(num_l, num_c):
  matriz = []
  for l in range(num_l): # inicia um laço for para preenchar linha por linha da matriz
    linha = []
    for c in range(num_c): # inicia outro laço for para preencher as colunas.
      elemento = int(input("Digite o elemento que será armazenado na posição {},{}: ".format(l, c))) # comando de entrada para receber cada elemento matricial.
      linha.append(elemento)
    matriz.append(linha) # adiciona na matriz as linhas que foram preenchidas.
  return matriz

num_linhas_mat_1 = int(input("Digite o número de linhas da primeira matriz: ")) # entrada com o número de linhas da primeira matriz.
num_colunas_mat_1 = int(input("Digite o número de colunas da primeira matriz: ")) # entrada com o número de colunas da primeira matriz.

matriz_1 = criar_matriz(num_linhas_mat_1, num_colunas_mat_1) # armaneza na variável matriz_1 a matriz de retorno resultante da função criar_matriz

print("A primeira matriz é: ", matriz_1) # imprime a primeira matriz

num_linhas_mat_2 = int(input("Digite o número de linhas da segunda matriz: ")) # entrada com o número de linhas da segunda matriz.
num_colunas_mat_2 = int(input("Digite o número de colunas da segunda matriz: ")) # entrada com o número de colunas da segunda matriz.

matriz_2 = criar_matriz(num_linhas_mat_2, num_colunas_mat_2) # armaneza na variável matriz_2 a matriz de retorno resultante da função criar_matriz.

print("A segunda matriz é: ", matriz_2) # imprime a segunda matriz.

if num_linhas_mat_1 == num_linhas_mat_2 and num_colunas_mat_1 == num_colunas_mat_2: # condição que verifica se as matrizes podem ser somadas.
  soma_matriz = np.add(matriz_1, matriz_2)
  print("A soma das matrizes resultou em: \n", soma_matriz)
else:
  print("A operação de soma não é possível, pois há uma incompatibilidade de dimensão matricial!!!")


if num_colunas_mat_1 == num_linhas_mat_2: # condição que verifica se as matrizes podem ser multiplicadas.
  prod_matriz = np.matmul(matriz_1, matriz_2)
  print("A multiplicação das matrizes resultou em: \n", prod_matriz)
else:
  print("A operação de multiplicação não é possível, pois há uma incompatibilidade de dimensão matricial!!!")
```
