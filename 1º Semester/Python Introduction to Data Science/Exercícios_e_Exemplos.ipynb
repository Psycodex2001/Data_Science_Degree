```python
import matplotlib.pyplot as plt
from matplotlib import cm
import numpy as np

X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
#Z = np.sin(R)
Z =X**2-Y**2

fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.plot_surface(X, Y, Z, cmap="Blues")

#ax.set(xticklabels=[], yticklabels=[], zticklabels=[])
ax.set(xlabel=("X"), ylabel=("Y"), zlabel=("Z"))
ax.zaxis.labelpad=1

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# constantes
a, b, x0, y0 = 3, 3, 5, 19
# variação do parâmetro t
t=np.linspace (-1, 2 * np.pi, 100)
#equações paramétricas
x=a*np.cos(t) + x0
y=b*np.sin(t) + y0

# faz o gráfico
plt.plot(x, y)
# variação dos eixos
#plt.axis([-6, 2, -3, 5])
# mostra o gráfico
plt.show()
```

```python
import matplotlib.pyplot as plt
# constantes

a, b, x0, y0 = 3, 2, -2, 1

# variação do parâmetro t
t=np.linspace (-1, 1, 100)

# equações paramétricas

x=a * np.cosh(t) + x0
y=b * np.sinh(t) + y0

# faz o gráfico

plt.plot( x, y)
plt.plot(-x, y, color='r')

sns.relplot(-x, y,kind="line",ci=None)

# variação dos eixos

plt.axis([-5, 5, -2, 4])

# mostra o gráfico

plt.show()
```

```python
import numpy as np
#constante
n = 8

# variação do parâmetro t

t=np.linspace (0, 2* np.pi, 500)

#paramétricas em coordenadas polares

theta = t

r = np.cos (n*t)

# paramétricas em coordenadas cartesianas

x = r * np.cos (theta)
y=r* np.sin(theta)

#fazendo o gráfico

plt.plot(x, y)

#mostrando o gráfico

plt.show()
```
